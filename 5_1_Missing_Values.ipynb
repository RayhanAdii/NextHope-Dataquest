```python
index = df_train[df_train['wind_spd'].isna()].index

# liat data yang missing sama intervalnya
missing_with_intervals = df_train[(df_train.index.isin(index)) | (df_train.index.isin(index-1)) | (df_train.index.isin(index+1))]
missing_with_intervals.head(12)
```

```python
# isi missing value dengan nilai rata-rata dari data sebelum dan sesudahnya
df_train['wind_spd'] = df_train['wind_spd'].fillna((df_train['wind_spd'].shift() + df_train['wind_spd'].shift(-1))/2)
```

```python
df_train['date'] = df_train['datetime_iso'].dt.date

# jd gweh ngegroup by date then cari mean-nyah
group_by_date = round(df_train.groupby('date')['wind_spd'].mean(), 2)
```

```python
# isi missing value dengan rata2 di dalam group_by_date
df_train['wind_spd'] = df_train['wind_spd'].fillna(df_train['date'].map(group_by_date))
```
